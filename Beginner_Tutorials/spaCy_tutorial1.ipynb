```python
matcher = Matcher(nlp.vocab)
pattern = [{"POS": "PROPN"}]
matcher.add("PROPER_NOUN",[pattern])
doc = nlp(text)
matches = matcher(doc)
print(len(matches))
for match in matches[:10]:
    print(match,doc[match[1]:match[2]])
```

The problem is that united and states are considered seperate proper nouns which is not the case.

```python
matcher = Matcher(nlp.vocab)
pattern = [{"POS": "PROPN",'OP':'+'}]
matcher.add("PROPER_NOUN",[pattern],greedy='LONGEST')
doc = nlp(text)
matches = matcher(doc)
print(len(matches))
for match in matches[:10]:
    print(match,doc[match[1]:match[2]])
```

The problem with greedy search is that the pnouns obtained are not ordered sequentially

```python
matcher = Matcher(nlp.vocab)
pattern = [{"POS": "PROPN",'OP':'+'}]
matcher.add("PROPER_NOUN",[pattern],greedy='LONGEST')
doc = nlp(text)
matches = matcher(doc)
matches.sort(key = lambda x: x[1])
print(len(matches))
for match in matches[:10]:
    print(match,doc[match[1]:match[2]])
```

If we want proper noun followed by a verb as well

```python
matcher = Matcher(nlp.vocab)
pattern = [{"POS": "PROPN",'OP':'+'},{"POS":"VERB"}]
matcher.add("PROPER_NOUNs",[pattern],greedy='LONGEST')
doc = nlp(text)
matches = matcher(doc)
matches.sort(key = lambda x: x[1])
print(len(matches))
for match in matches[:10]:
    print(match,doc[match[1]:match[2]])
```

*Extract text and match after a comma*

```python
matcher = Matcher(nlp.vocab)
pattern = [{"ORTH": ","},
           {"IS_ALPHA":True,'OP':"+"}, # + -> will occur
           {'IS_PUNCT':True,"OP":"*"} # * -> may occur
           ]
matcher.add("PROPER_NOUNs",[pattern],greedy='LONGEST')
doc = nlp(text)
matches = matcher(doc)
#matches.sort(key = lambda x: x[1])
print(len(matches))
for match in matches[:10]:
    print(match,doc[match[1]:match[2]])
```

Adding people speaking and speaking to as we expect around the comma, we get the person speaking and the person spoken to.

```python
speak_lemmas = ['feel','reach']
matcher = Matcher(nlp.vocab)
pattern = [{"ORTH": ","},
           {"IS_ALPHA":True,'OP':"+"}, # + -> will occur
           {'IS_PUNCT':True,"OP":"*"}, # * -> may occur
           {'POS':'VERB','LEMMA':{'IN': speak_lemmas}},
           {'POS':'PROPN','OP':'+'},
           {'ORTH':','}
           ]
matcher.add("PROPER_NOUNs",[pattern],greedy='LONGEST')
doc = nlp(text)
matches = matcher(doc)
#matches.sort(key = lambda x: x[1])
print(len(matches))
for match in matches[:10]:
    print(match,doc[match[1]:match[2]])
```

```python
@Language.component('remove_gpe')
def remove_gpe(doc):
    orignal_ents = list(doc.ents)
    for ent in doc.ents:
        if ent.label == 'GPE':
            orignal_ents.remove(ent)
    doc.ents = orignal_ents
    return doc
```

```python
from spacy.language import Language

@Language.component('bradley_ner')
def bradley_ner(doc):
    pattern = r"Bradley [A-Z]\w+"
    orignal_ents = list(doc.ents)
    mwt_ents = []
    for match in re.finditer(pattern, doc.text):
        start, end = match.span()
        span = doc.char_span(start, end)
        if span is not None:
            mwt_ents.append((span.start,span.end, span.text))

    for ent in mwt_ents:
        start, end, name = ent
        per_ent = Span(doc, start, end, label = "PERSON")
        orignal_ents.append(per_ent)

    doc.ents =orignal_ents
    return (doc)
```
